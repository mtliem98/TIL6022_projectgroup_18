**Objectives:**

Modelling the effect of building the Lelylijn on line usage in the surrounding area. 


**Research questions:**

How many people travel between each station currently?

How many passengers will use the Lelylijn without any additional demand?

How does the passenger flow change with the Lelylijn on the surrounding train lines without any additional demand?

See how the newly built line would affect the travel time map. How many people have better access to big transport hubs?

Try to predict the number of new public transport users along Lelylijn. What will be the additional demand because of the Lelylijn? (if we have extra time)


**Intended data pipeline analysis:**

Import data on all stations (also the new ones already or later?) and lines.

Assign travel times to the lines.

Determine the number of passengers boarding/leaving the train at each station.

Try to predict the general travel directions based on data used in previous steps.

Assign the passenger flows onto the lines. Create a matrix from the data.

Calculate the projected change in travel times (and create a map on it).

Use a shortest path algorithm to determine the passenger numbers on the Lelylijn and the other lines in the region.

Create more charts/maps on our findings.

*note that things with a/b assigned can be done in parallel
